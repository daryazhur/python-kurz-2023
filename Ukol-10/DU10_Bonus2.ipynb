```python
import pandas
nepracujici = pandas.read_csv('nepracujici.csv')
vykazy = pandas.read_csv('vykazy.csv')

vykazy = vykazy.rename(columns = {'date' : 'datum', 'hours':'casy', 'project':'projekt', 'emloyee_id':'cislo_zamestnance'})
```

* Propoj tabulku s výkazy s tabulkou se zaměstnanci, kterou jsi vytvořil(a) v předchozím cvičení.

```python
vykazy_nepracujici = pandas.merge(vykazy, nepracujici, on = ['cislo_zamestnance'])
vykazy_nepracujici
```

* Následně proveď statistiku vykázaných hodin za jednotlivé kanceláře, tj. spočítej celkový počet hodin vykázaný zaměstnanci jednotlivých kanceláří na projekty daného zákazníka.

```python
vykazy_nepracujici.groupby('projekt')['casy'].count()
```
